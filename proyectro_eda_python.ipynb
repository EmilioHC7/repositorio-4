```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
#INFORMACION OBTENIDA DEL BANCO

df_bank_csv = pd.read_csv("/Users/emilio/Desktop/DatosProyecto/bank-additional.csv")
df_bank_csv
```

```python
# Características demográficas y comportamiento de compra de los clientes del banco.
# Cargar las tres hojas del Excel en un solo de DataFrames

df_customer_excel = pd.read_excel("/Users/emilio/Desktop/DatosProyecto/customer-details.xlsx", sheet_name=None)
df_customer_excel
```

```python
# Unir todas las hojas en un solo DataFrame
df_excel = pd.concat(df_customer_excel.values(), ignore_index=True)
df_excel
```

```python
# Limpieza de datos documento csv

df_bank_csv.drop_duplicates(inplace=True)
df_bank_csv
```

```python
#Remplazar los valores nulos en las columnas por "desconocido" 

df_bank_csv.fillna("desconocido", inplace=True)
df_bank_csv
```

```python
#Analisis estadistico basico

print(df_bank_csv.describe(include='all'))
```

```python
# ahora con los datos de caracteristicas demograficas del cliente
print(df_excel.describe(include='all'))
```

```python
#Verificar valores nulos
print(df_bank_csv.isnull().sum())
```

```python
print(df_excel.isnull().sum())
```

```python
#Comprobamos si hay claves en comun

common_ids = set(df_bank_csv['id_']) & set(df_excel['ID'])
print(f"IDs en comun: {len(common_ids)}")
```

```python
#Verificacion de que no haya duplicados en la columna ID en ambas tablas

# Primero la tabla con los datos del banco 'id_'
print(df_bank_csv['id_'].duplicated().sum())
```

```python
# ahora comprobamos que no haya duplicados en la columna ID de tabla customer
print(df_excel['ID'].duplicated().sum())
```

```python
from pandas.core.reshape.merge import merge

df_bank_csv.rename(columns={'id_':'ID'}, inplace=True)
```

```python
# Teniendo en cuenta que ambas tablas comporten relacion por medio de la columna id_ y ID, haremos un merge para generar un nuevo DataFrame

df_final = df_bank_csv.merge(df_excel, on='ID', how='inner')
df_final
```

```python
# Asignamos el ID como indice
df_final.set_index('ID', inplace=True)
df_final
```

```python
# Calcular la tasa de conversion de la campaña de marketing

conversion_rate = df_final['y'].value_counts(normalize=True) * 100
conversion_rate
```

```python
# Relacion entre duracion de la llamada, numero de  contactos y conversion

duracion_conversion = df_final.groupby('y')['duration'].mean()
campaña_conversion = df_final.groupby('y')['campaign'].mean()
```

```python
duracion_conversion # en segundos
```

```python
campaña_conversion # numero de veces contactados
```

```python
# Caracteristicas de clientes asociados a una respuesta positiva

positive_customer = df_final[df_final['y'] == 'yes'].describe()
positive_customer
```

```python
# Metodos de contacto mas efectivos

contact_conversion = df_final.groupby('contact')['y'].value_counts(normalize=True).unstack()*100
contact_conversion

#Podemos notar que el metodo mas efectivo fue el celular con un casi 15% de conversion en comparacion de llamada telefonica de un 5%
```

```python
# Relacion entre antiguedad del cliente y conversion

df_final['customer_age']= (pd.to_datetime('2025-01-01')- df_final['Dt_Customer']).dt.days
cliente_año_coversion = df_final.groupby('y')['customer_age'].mean()
cliente_año_coversion
```
